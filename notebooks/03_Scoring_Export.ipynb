```python
import pandas as pd

# Charger le dataset clean
df = pd.read_csv('/content/clean_churn_dataset.csv')

# Prédire avec la régression logistique
# ⚠️ Tu dois recharger le modèle ou le recréer ici
from sklearn.linear_model import LogisticRegression

# Créer et réentraîner la régression logistique
X = df.drop('Exited', axis=1)
y = df['Exited']

logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X, y)

# Ajouter les prédictions
df["Predicted_Churn"] = logreg.predict(X)
df["Churn_Probability"] = logreg.predict_proba(X)[:,1]

# Vérifier
df.head()
```

```python
df.to_csv('/content/churn_scoring_dataset.csv', index=False)
```
